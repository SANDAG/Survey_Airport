### Complete SP

```python
df = pd.read_csv('../data/processed/atc_travel_survey_final_data.csv')
df['respondent_segment_label'] = np.where(df['marketsegment_label']=='EMPLOYEE', 'EMPLOYEE', df['passenger_segment_label'])
#working_df = df[(df['is_completed'] == True) & (df['record_type_synthetic'] == False)]
working_df = df[(df['record_type_synthetic'] == False)]

working_df['sp_complete'] = (
    ~working_df['sp_connection_to_old_town_center'].isna() |
    ~(working_df['sp_other_airport_list'].isna() | (working_df['sp_other_airport_list'].astype(str).str.strip() == ''))
)
```

```python
working_df.groupby(['respondent_segment_label', 'is_self_administered', 'is_pilot', 'sp_complete']).size().reset_index(name = 'Total number')
```
