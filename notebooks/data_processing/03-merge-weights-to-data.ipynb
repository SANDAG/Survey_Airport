```python
import pandas as pd
```

```python
data_path = '../../data'
data_model_output = pd.read_csv(f'{data_path}/processed/data_model_output.csv')
weights_departing_and_arriving = pd.read_csv(f'{data_path}/interim/survey_weights_only_departing_and_arriving.csv')
weights_departing_only = pd.read_csv(f'{data_path}/interim/survey_weights_only_departing_only.csv')
weights_departing_only_model_respondents = pd.read_csv(f'{data_path}/interim/survey_weights_departing_only_model_respondents.csv')
weights_non_sas_departing_only = pd.read_csv(f'{data_path}/interim/survey_weights_only_departing_non_sas_only.csv')
weights_departing_only_with_time_of_day = pd.read_csv(f'{data_path}/interim/survey_weights_only_departing_only_with_time_of_day.csv')
```

```python
# Drop the original weight column if it exists
data_model_output = data_model_output.drop(columns=['weight'], errors='ignore')

# Overwrite existing columns with new weight values using unique_id
for df, col in [
    (weights_departing_and_arriving, 'weight_departing_and_arriving'),
    (weights_departing_only, 'weight_departing_only'),
    (weights_departing_only_model_respondents, 'weight_departing_only_model_respondents'),
    (weights_non_sas_departing_only, 'weight_non_sas_departing_only'),
    (weights_departing_only_with_time_of_day, 'weight_departing_only_with_time_of_day'),
]:
    data_model_output = data_model_output.drop(columns=[col], errors='ignore')  # drop existing col to avoid ambiguity
    data_model_output = data_model_output.merge(
        df[['unique_id', 'weight']].rename(columns={'weight': col}),
        on='unique_id', how='left'
    )
```

```python
data_model_output.to_csv(f'{data_path}/processed/data_model_output.csv', index=False)
```
